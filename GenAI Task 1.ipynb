```python
'''Handwritten Digit Classification using MLP (MNIST)

Step 1 – Import Libraries'''
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
import matplotlib.pyplot as plt
import numpy as np
```

```python
'''Step 2 – Load the MNIST Dataset'''
# Load MNIST dataset

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

print("Training Images :", x_train.shape)
print("Training Labels :", y_train.shape)

print("Testing Images  :", x_test.shape)
print("Testing Labels  :", y_test.shape)
```

```python
'''Step 3 – Display the First Image'''
plt.imshow(x_train[0], cmap="gray")
plt.title(f"Label : {y_train[0]}")
plt.axis("off")
plt.show()
```

```python
'''Step 4 – Normalize the Images'''
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
'''Step 5 – Build the MLP'''
model = Sequential([
    
    Flatten(input_shape=(28,28)),
    
    Dense(128, activation="relu"),
    
    Dense(10, activation="softmax")
])
```

```python
''' step 5: explanation:
28 × 28 Image
↓
Flatten
↓
784 Pixels
↓
128 Neurons
↓
10 Output Neurons
↓
Digit Prediction'''
```

```python
'''Step 6 – Compile the Model'''
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
```

```python
'''Step 7 – Train the Model'''
history = model.fit(
    x_train,
    y_train,
    epochs=5,
    verbose=1
)
```

```python
'''Step 8 – Evaluate the Model'''
loss, accuracy = model.evaluate(x_test, y_test)

print("Test Loss :", loss)
print("Test Accuracy :", accuracy)
```

```python
'''Step 9 – Display the Input Image'''

# Select any test image
index = 25

# Display the image
plt.imshow(x_test[index], cmap="gray")
plt.title("Input Image")
plt.axis("off")
plt.show()

print("Actual Digit :", y_test[index])
```

```python
'''Step 10 – Predict the Digit'''

# Predict the selected image
prediction = model.predict(x_test[index:index+1])

predicted_digit = np.argmax(prediction)

print("Predicted Digit :", predicted_digit)
```

```python
'''Method 2 - Predict a Specific Digit:
Step 1: Find the First Image of a Specific Digit'''

import numpy as np
import matplotlib.pyplot as plt

# Choose the digit you want
digit = 7

# Find the first occurrence of digit 7
index = np.where(y_test == digit)[0][0]

# Display the image
plt.imshow(x_test[index], cmap="gray")
plt.title(f"Actual Digit: {y_test[index]}")
plt.axis("off")
plt.show()
```

```python
'''Step 2: Predict the specific input image'''

prediction = model.predict(x_test[index:index+1])

predicted_digit = np.argmax(prediction)

print("Predicted Digit:", predicted_digit)
print("Actual Digit   :", y_test[index])
```
